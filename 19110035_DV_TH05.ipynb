# Bài tập

Vẽ choropleth map cho dataset  number-of-internet-users-by-country.csv

```python
import pandas as pd
name = 'number-of-internet-users-by-country.csv'

internet_usage_df = pd.read_csv(name)
internet_usage_df = internet_usage_df.sort_values(by=['Year'])
internet_usage_df.head()
```

```python
#Let’s generate the animated plot again now that the sorting is done
import plotly.express as px
fig = px.scatter_geo(internet_usage_df, locations="Code",
                    size="Number of internet users (OWID based on WB & UN)", # lifeExp is a column of gapminder
                    hover_name="Entity", # column to add to hover information
                    animation_frame="Year", # column on which to animate
)
                    
fig.update_layout(
    # add a title text for the plot
    title_text = 'Internet usage across the world (% population)',
    # set projection style for the plot
    geo = dict(projection={'type':'natural earth'}) # by default, projection type is set to 'equirectangular'
)

fig.show()
```
